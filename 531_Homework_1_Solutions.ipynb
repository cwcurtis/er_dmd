```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Problem 1

So our characteristic equations are
$$
\frac{dx}{dt} = -3, ~ \frac{du}{dt} = 0,
$$
so we have 
$$
x=-3t + x_{0}, ~ u(x(t),t) = u_{0}(x_{0}) = \cos(x_{0}),
$$
and thus $u(x,t)=\cos(x+3t)$.

To visualize this, we can try plotting $u(x,t)$ for several different times.  In the following figure, we look at $u(x,0)$, $u(x,1)$, and $u(x,5)$.

```python
Npts = 1000
xvals = np.linspace(-20., 20., Npts)
tvals = [0., 1., 5.]
colors = ['k', 'r', 'b']
labels = ['u(x,0)', 'u(x,1)', 'u(x,5)']
```

```python
for jj, tval in enumerate(tvals):
    plt.plot(xvals, np.cos(xvals  + 3.*tval),c=colors[jj], label=labels[jj])
plt.legend()
plt.xlabel("x")
```

Okay, we kind of see how the wave propagates from right to left.  That said, we get a much better idea of the dynamics when we look at $u(x,t)=\cos(x+3t)$ superimposed over the characteristics in the $x-t$ plane.

```python
Nstps = 100
tvals_fine = np.linspace(0., 2., Nstps)
uvals = np.zeros((xvals.size, tvals_fine.size))

# Plot the characteristics
x0s = np.arange(-20,30)
for x0 in x0s:
    plt.plot(xvals, -(xvals-x0)/3., color='k', ls='--')
    
# Plot u(x,t) over the characteristics
for jj in range(tvals_fine.size):
    uvals[:, jj] = np.cos(xvals + 3.*tvals_fine[jj])
plt.pcolor(xvals, tvals_fine, uvals.T)
plt.ylim([np.min(tvals_fine), np.max(tvals_fine)])
plt.colorbar()
plt.xlabel("x")
plt.ylabel("t");
```

## Problem 2

So our characteristic equations become
$$
\frac{dx}{dt} = c, ~ \frac{du}{dt} = e^{-t},
$$
which integrate to 
$$
x = ct + x_{0}, ~ u(x(t), t) = 1 - e^{-t} + u_{0}(x_{0})
$$
so we finally have 
$$
u(x,t) = 1 - e^{-t} + u_{0}(x-ct).
$$
The characteristics are just parallel lines in the $x-t$ plane.  


While I didn't ask you to illustrate the following, it is instructive to see it anyway.  So, as we can see, since 
$$
\lim_{t\rightarrow \infty} e^{-t} = 0
$$
as $t\rightarrow \infty$, we see our solution $u(x,t)$ goes like
$$
u(x,t) \sim 1 + u_{0}(x-ct), ~ t \rightarrow \infty.
$$

Thus, our solution tends to just propagate the initial condition along characteristics while the overall level of the solution comes up by exactly one.  To see this, suppose we take $u_{0}(x)=\tanh(x)$ and then compare various values of $u(x,t)$ to see the effect as shown in the figure below.  

```python
tvals_coarse = [0., 2., 4.]
pb2labels = ['u(x,0)', 'u(x,2)', 'u(x,4)']
for jj, tval in enumerate(tvals_coarse):
    plt.plot(xvals, np.tanh(xvals - tval) + 1. - np.exp(-tval),c=colors[jj],label=pb2labels[jj])
plt.xlabel("x")
plt.legend()
```

## Problem 3

So our characteristic equations become
$$
\frac{dx}{dt} = x, ~ \frac{du}{dt} = 1,
$$
which integrate to 
$$
x = x_{0}e^{t}, ~ u(x(t), t) = t + u_{0}(x_{0})
$$
so we finally have 
$$
u(x,t) = t + \tanh(xe^{-t}),
$$
which, given that 
$$
\lim_{t\rightarrow \infty}xe^{-t} = 0,
$$
and $\tanh(0)=0$, we see that 
$$
u(x,t) \sim t, ~ t\rightarrow \infty.
$$

To see all of this, we first look at the characteristics, and how the **initial condition** propagates over them.  This is shown in the following figure.  

```python
Nstps = 100
tvals_fine = np.linspace(0., 2., Nstps)
uvals = np.zeros((xvals.size, tvals_fine.size))

# Plot the characteristics
x0s = np.arange(-20,20)+.01
for x0 in x0s:
    plt.plot(xvals, np.log(np.abs(xvals/x0)), color='k', ls='--')
    
# Plot u(x,t) over the characteristics
for jj in range(tvals_fine.size):
    uvals[:, jj] = np.tanh(xvals*np.exp(-tvals_fine[jj]))
plt.pcolor(xvals, tvals_fine, uvals.T)
plt.ylim([np.min(tvals_fine), np.max(tvals_fine)])
plt.colorbar()
plt.xlabel("x")
plt.ylabel("t");
```

From this, we see that the values of the initial condition are pushed away from one another, leaving only $u=0$ to remain.  Likewise, we can plot several values of $u(x,t)$ to get a sense of what is going on.  

```python
tvals_coarse = [0., 1., 2.]
pb3labels = ['u(x,0)', 'u(x,1)', 'u(x,2)']
for jj, tval in enumerate(tvals_coarse):
    plt.plot(xvals, np.tanh(xvals*np.exp(-tval)) + tval,c=colors[jj],label=pb3labels[jj])
plt.xlabel("x")
plt.legend()
```

Thus, we see that as $t$ increases, $u(x,t)$ flattens and rises, as we showed above.  

## Problem 4

So our characteristic equations are
$$
\frac{dx}{dt} = u(x(t),t), ~ \frac{du}{dt} = 0
$$

so that $u(x(t),t)=u_{0}(x_{0})$.  We then have that 
$$
x = u_{0}(x_{0})t + x_{0}
$$
so for $u_{0}(x_{0})=1$, we have $x=t+x_{0}$ and for $u_{0}(x_{0})=0$, we have $x=x_{0}$.  

We get a shock starting at $(x,t)=(0,0)$, so we need to find the corresponding shock path $x_{s}(t)$ using the Rankine--Huginot condition.  In this case, we see from $f(u)=u^{2}$ that $F(u)=u^{3}/3$, and thus 
$$
\frac{dx_{s}}{dt} = \frac{1}{3}\left(\frac{u_{+}^{3}-u_{-}^{3}}{u_{+}-u_{-}}\right) = \frac{1}{3}\left(u_{+}^{2}+u_{+}u_{-} + u_{-}^{2}\right) = \frac{1}{3}
$$
given that $u_{+}=0$ and $u_{-}=1$.  Thus we have $x_{s}(t)=t/3$.  Putting it all together, we get
$$
u(x,t) = \left\{\begin{array}{rl} 1 & x < t/3 \\ 0 & x \geq t/3\end{array}\right.
$$


To see all of this put together, we first go to the $(x,t)$ plane and plot characteristics.  

```python
Nstps = 100
Npts = 1000

tvals_fine = np.linspace(0., 2., Nstps)
xvals = np.linspace(-2., 1., Npts)

# Plot the characteristics
x0s = np.linspace(-3,3,40)
for x0 in x0s:
    if x0 < 0:
        indskp = xvals < np.abs(x0)/2.
        plt.plot(xvals[indskp], xvals[indskp]-x0, color='k', ls='--')
    else:
        plt.plot([x0, x0], [np.min(tvals_fine), 3.*x0], color='b', ls='--')
        
plt.plot(xvals,3*xvals, color='r', linewidth=4.)
plt.ylim([np.min(tvals_fine), np.max(tvals_fine)])
plt.xlim([-1, 2])
plt.xlabel("x")
plt.ylabel("t")
```

## Problem 5

So again, our characteristic equations are
$$
\frac{dx}{dt} = u(x(t),t), ~ \frac{du}{dt} = 0
$$

so that $u(x(t),t)=u_{0}(x_{0})$.  We then have that 
$$
x = u_{0}(x_{0})t + x_{0}
$$
so for $u_{0}(x_{0})=1$, we have $x=t+x_{0}$ and for $u_{0}(x_{0})=0$, we have $x=x_{0}$.  

This time, we get a rarefaction region, so we need to find the corresponding similarity solution $u(x,t)=g(x/t)$.  We let $z=x/t$ so that our PDE becomes
$$
-\frac{z}{t}g'(z) + \frac{1}{t}g^{2}(z)g'(z) = 0.
$$
For $t>0$, we then find that the interesting solution relevant to our interests is given by
$$
g(z)=\sqrt{z}, ~\text{or} ~u(x,t) = \sqrt{x/t}.
$$

Putting it all together, we get
$$
u(x,t) = \left\{\begin{array}{rl} 0 & x < 0 \\ \sqrt{x/t} & 0 \leq x < t \\ 1 & x\geq t\end{array}\right.
$$

```python
Nstps = 100
Npts = 1000

tvals_fine = np.linspace(0., 2., Nstps)
xvals = np.linspace(-2., 4., Npts)

# Plot the characteristics
x0s = np.linspace(-2,4,40)
for x0 in x0s:
    if x0 >= 0:
        plt.plot(xvals, xvals-x0, color='k', ls='--')
    else:
        plt.plot([x0, x0], [np.min(tvals_fine), np.max(tvals_fine)], color='b', ls='--')

theta_vals = np.linspace(np.pi/4, np.pi/2, 20)
for theta in theta_vals:
    plt.plot(xvals, np.tan(theta)*xvals, color='r')

plt.ylim([np.min(tvals_fine), np.max(tvals_fine)])
plt.xlim([-1, 2])
plt.xlabel("x")
plt.ylabel("t")
```

```python
def usol(x,t):
    if x < 0:
        return 0
    elif 0 <= x and x < t:
        return np.sqrt(x/t)
    else:
        return 1
```

```python
tvals_coarse = [0., 2., 4.]
xvals = np.linspace(-2., 8., Npts)
pb2labels = ['u(x,0)', 'u(x,2)', 'u(x,4)']
for jj, tval in enumerate(tvals_coarse):
    utmp = np.zeros(xvals.size)
    for cnt, xval in enumerate(xvals):
        utmp[cnt] = usol(xval, tval)
    plt.plot(xvals, utmp,c=colors[jj],label=pb2labels[jj])
plt.xlabel("x")
plt.legend()
```
